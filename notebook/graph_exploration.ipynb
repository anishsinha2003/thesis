```python
import scanpy as sc

# Load processed dataset
adata = sc.read_h5ad(
    "../data/processed/GSE190075_sample_1.h5ad"
)

# ============================================================
# Dataset Overview
# ============================================================

# Overall AnnData object
# Shows number of cells, genes and stored information
print("DATASET SUMMARY")
print(adata)

# Expression matrix
# Rows = cells
# Columns = genes
# Values = normalized gene expression
print("\nCELL x GENE MATRIX")
print(adata.X.shape)

# ============================================================
# Cell Information
# ============================================================

# Metadata for each cell
# Each row represents one cell
print("\nCELL METADATA")
print(adata.obs.head())

# ============================================================
# Gene Information
# ============================================================

# Metadata for each gene
# Each row represents one gene
print("\nGENE METADATA")
print(adata.var.head())

# ============================================================
# Cell Features
# ============================================================

# PCA embeddings for each cell
# These are likely candidates for cell node features
print("\nPCA EMBEDDINGS")
print(adata.obsm.keys())

print("\nPCA SHAPE")
print(adata.obsm["X_pca"].shape)

print("\nFIRST 5 CELL EMBEDDINGS")
print(adata.obsm["X_pca"][:5])

# ============================================================
# Cell-Gene Relationships
# ============================================================

# Expression values connect cells to genes
# Could become Cell -> Gene edges later

print("\nFIRST 5 CELLS x FIRST 5 GENES")

if hasattr(adata.X, "toarray"):
    print(adata.X[:5, :5].toarray())
else:
    print(adata.X[:5, :5])

# ============================================================
# Cell-Cell Relationships
# ============================================================

# Generated by Scanpy kNN graph
# Could become Cell -> Cell edges

print("\nGRAPH MATRICES")
print(adata.obsp.keys())

print("\nCONNECTIVITY MATRIX")
print(adata.obsp["connectivities"])

# ============================================================
# Thesis Graph Mapping
# ============================================================

print("\nGRAPH DESIGN")

print("Cell nodes      -> adata.obs")
print("Gene nodes      -> adata.var")
print("Cell features   -> adata.obsm['X_pca']")
print("Cell-Gene edges -> adata.X")
print("Cell-Cell edges -> adata.obsp['connectivities']")
```

```python
# checks
# ============================================================
# PREPROCESSING VALIDATION CHECKS
# ============================================================

print("===================================")
print("1. DATASET SHAPE")
print("===================================")

print(adata)
print()

# Expect:
# - Thousands of cells
# - Around 2000 genes after HVG selection


print("===================================")
print("2. CELL QC CHECK")
print("===================================")

print(adata.obs["n_genes_by_counts"].describe())
print()

# Check:
# min should be >= 200
# because we filtered cells with min_genes=200


print("===================================")
print("3. NUMBER OF GENES AFTER HVG")
print("===================================")

print("Genes remaining:", adata.n_vars)
print()

# Expect:
# ~2000 genes if HVG selection worked


print("===================================")
print("4. PCA EXISTS")
print("===================================")

print(adata.obsm.keys())
print()

# Expect:
# 'X_pca'


print("===================================")
print("5. PCA SHAPE")
print("===================================")

print(adata.obsm["X_pca"].shape)
print()

# Expect:
# (number_of_cells, ~50)


print("===================================")
print("6. PCA VALUES")
print("===================================")

print(adata.obsm["X_pca"][:5])
print()

# First 5 cell embeddings


print("===================================")
print("7. CELL METADATA")
print("===================================")

print(adata.obs.head())
print()

# Metadata stored for each cell


print("===================================")
print("8. GENE METADATA")
print("===================================")

print(adata.var.head())
print()

# Metadata stored for each gene


print("===================================")
print("9. EXPRESSION VALUES")
print("===================================")

if hasattr(adata.X, "toarray"):
    print(adata.X[:5, :5].toarray())
else:
    print(adata.X[:5, :5])

print()

# Values should be:
# normalized + log-transformed
# NOT huge raw counts


print("===================================")
print("10. BASIC VALIDATION")
print("===================================")

print("Cells:", adata.n_obs)
print("Genes:", adata.n_vars)

if "X_pca" in adata.obsm:
    print("PASS: PCA exists")
else:
    print("FAIL: PCA missing")

if adata.n_vars <= 2500:
    print("PASS: HVG selection likely worked")
else:
    print("WARNING: Too many genes remain")

if adata.obs["n_genes_by_counts"].min() >= 200:
    print("PASS: QC filtering worked")
else:
    print("FAIL: Cells with <200 genes still exist")
```
